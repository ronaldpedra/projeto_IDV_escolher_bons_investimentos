```python
import pandas as pd
import numpy as np
```

```python
# Configurações
caminho_planilha = './planilhas/Planilha de ouro FIIs - Abril 2024.xlsx'

# Variáveis
administrador_desejavel = ['Credit Suisse Hedging-Griffo', 'INTRAG DTVM', 'Banco Fator', 'BRL Trust', 'COINVALORES', 'XP Investimentos',
                           'Rio Bravo Investimentos']

administrador_aceitavel = ['BTG Pactual', 'Votorantim Asset Management', 'CM Capital Markets', 'Banco J. Safra']
administrador_aceitavel = set(administrador_desejavel + administrador_aceitavel)
administrador_aceitavel = list(administrador_aceitavel)

administrador_ruim = ['Oliveira Trust', 'Vórtx DTVM LTDA', 'Hedge Investments DTVM LTDA', 'Santander Securities Services Brasil DTVM', 
                      'Genial Investimentos Corretora de Valores']
```

```python
def converte_coluna_float(df_original, lista_nome_coluna):

    for nome_coluna in lista_nome_coluna:
        df_original[nome_coluna] = pd.to_numeric(df_original[nome_coluna], errors='coerce')
        df_original[nome_coluna] = df_original[nome_coluna].replace(np.NaN, 0)

    return df_original
```

```python
tabela = pd.read_excel(caminho_planilha, skiprows=2)
```

```python
colunas_float = ['Liquidez diária', 'Patrimônio Líquido', 'Taxa de Performance', '% em CRIs', 'Quantidade de CRIs']
tabela = converte_coluna_float(tabela, colunas_float)

tabela['Benchmark'] = tabela['Benchmark'].replace(np.NaN, "-")
tabela
```

```python
# LOGÍSTICA
# FIIs de Logística desejáveis
filtro_logistica_desejavel = tabela[
    (tabela['Tipo de fundo'].isin(['Tijolo', 'Papel'])) 
    & (tabela['Segmento de Atuação'] == 'Logística') 
    & (tabela['Quantidade de Imóveis'] >= 10) 
    & (tabela['Multi-inquilino'] == 'Sim') 
    & (tabela['Vacância'] <= 0.25) 
    & (tabela['Tempo de listagem'] >= 5)
    & (tabela['Administrador'].isin(administrador_desejavel))

    & ((tabela['Tipo de Gestão'] == 'Passiva') & (tabela['Taxa de adm'] <= 0.002) & (tabela['Taxa de Performance'] == 0) | 
       (tabela['Tipo de Gestão'] == 'Ativa') & (tabela['Taxa de adm'] <= 0.02) & (tabela['Taxa de Performance'] <= 0.2))

    & ((tabela["Benchmark"].str.contains("-")) | (tabela["Benchmark"].isnull()) | (tabela["Benchmark"].str.contains("IPCA", case=False)))
]

# FIIs de Logística aceitáveis
filtro_logistica_aceitavel = tabela[
    (tabela['Tipo de fundo'].isin(['Tijolo', 'Papel', 'Hibrido'])) 
    & (tabela['Segmento de Atuação'] == 'Logística') 
    & (tabela['Quantidade de Imóveis'] >= 6)
    & (tabela['Vacância'] <= 0.3) 
    & (tabela['Tempo de listagem'] >= 3)
    & (tabela['Administrador'].isin(administrador_aceitavel))

    & ((tabela['Tipo de Gestão'] == 'Passiva') & (tabela['Taxa de adm'] <= 0.002) & (tabela['Taxa de Performance'] == 0) | 
       (tabela['Tipo de Gestão'] == 'Ativa') & (tabela['Taxa de adm'] <= 0.02) & (tabela['Taxa de Performance'] <= 0.2))

    & ((tabela["Benchmark"].str.contains("-")) | (tabela["Benchmark"].str.contains("IPCA", case=False)))
]
# Removendo as linhas do Dataframe aceitavel com o mesmo índice do DataFrame desejavel
filtro_logistica_aceitavel = filtro_logistica_aceitavel.drop(filtro_logistica_desejavel.index)

# FIIs de Logística Ruins
filtro_logistica_ruim = tabela[tabela['Segmento de Atuação'] == 'Logística']
# Removendo as linhas do Dataframe ruim com o mesmo índice do DataFrame desejavel e aceitavel
filtro_logistica_ruim = filtro_logistica_ruim.drop(filtro_logistica_desejavel.index)
filtro_logistica_ruim = filtro_logistica_ruim.drop(filtro_logistica_aceitavel.index)


# Cria os dataframes de logística
df_logistica_desejavel = filtro_logistica_desejavel.copy()
df_logistica_aceitavel = filtro_logistica_aceitavel.copy()
df_logistica_ruim = filtro_logistica_ruim.copy()


# Cria a coluna de classificação dos fii
df_logistica_desejavel['Classificação'] = 'logistica_desejavel'
df_logistica_aceitavel['Classificação'] = 'logistica_aceitavel'
df_logistica_ruim['Classificação'] = 'logistica_ruim'

df_logistica = pd.concat([df_logistica_desejavel, df_logistica_aceitavel, df_logistica_ruim], ignore_index=False)

display(df_logistica)
```

```python
# Títulos e Valores Mobiliários
# FIIs de Títulos e Val. Mob. desejáveis
filtro_tvm_desejavel = tabela[
    (tabela['Tipo de fundo'].isin(['Papel'])) 
    & (tabela['Segmento de Atuação'] == 'Títulos e Val. Mob.') 
    & (tabela['% em CRIs'] >= 0.75)
    & (tabela['Quantidade de CRIs'] >= 50)
    
    & (tabela['Liquidez diária'] >= 7000)
    & (tabela['Patrimônio Líquido'] >= 300000000)

    & (tabela['Tempo de listagem'] >= 5)
    & (tabela['Administrador'].isin(administrador_desejavel))

    & (tabela['Taxa de adm'] <= 0.01) & (tabela['Taxa de Performance'] == 0)

    #& ((tabela["Benchmark"].str.contains("-")) | (tabela["Benchmark"].isnull()) | (tabela["Benchmark"].str.contains("IPCA", case=False)))
]

# FIIs de Títulos e Val. Mob. aceitáveis
filtro_tvm_aceitavel = tabela[
    (tabela['Tipo de fundo'].isin(['Papel'])) 
    & (tabela['Segmento de Atuação'].isin(['Títulos e Val. Mob.', 'Híbrido']))

    & (tabela['Liquidez diária'] >= 7000)
    & (tabela['Patrimônio Líquido'] >= 300000000)

    & (tabela['% em CRIs'] >= 0.5)
    & (tabela['Quantidade de CRIs'] >= 40)
    

    & (tabela['Tempo de listagem'] >= 3)
    & (tabela['Administrador'].isin(administrador_aceitavel))

    & (tabela['Taxa de adm'] < 0.012) & (tabela['Taxa de Performance'] == 0)

    #& ((tabela["Benchmark"].str.contains("-")) | (tabela["Benchmark"].str.contains("IPCA", case=False)))
]
# Removendo as linhas do Dataframe aceitavel com o mesmo índice do DataFrame desejavel
filtro_tvm_aceitavel = filtro_tvm_aceitavel.drop(filtro_tvm_desejavel.index)


# FIIs de Títulos e Val. Mob. Ruins
filtro_tvm_ruim = tabela[tabela['Tipo de fundo'] == 'Papel']
# Removendo as linhas do Dataframe ruim com o mesmo índice do DataFrame desejavel e aceitavel
filtro_tvm_ruim = filtro_tvm_ruim.drop(filtro_tvm_desejavel.index)
filtro_tvm_ruim = filtro_tvm_ruim.drop(filtro_tvm_aceitavel.index)


# Criando os dataframes de tvm
df_tvm_desejavel = filtro_tvm_desejavel.copy()
df_tvm_aceitavel = filtro_tvm_aceitavel.copy()
df_tvm_ruim = filtro_tvm_ruim.copy()


# Cria a coluna de classificação dos fii
df_tvm_desejavel['Classificação'] = 'tvm_desejavel'
df_tvm_aceitavel['Classificação'] = 'tvm_aceitavel'
df_tvm_ruim['Classificação'] = 'tvm_ruim'

df_tvm = pd.concat([df_tvm_desejavel, df_tvm_aceitavel, df_tvm_ruim], ignore_index=False)

display(df_tvm)
```
